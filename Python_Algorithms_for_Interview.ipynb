## Big O

```python
def sum1(num):
  result = 0
  for x in range(num+1): #O(N)
    result += x
  return result
```

```python
sum1(15)
```

```python
def sum2(num):
  return (num*(num+1))//2
```

```python
sum2(15)
```

## Big(O) Examples 

### O(1) Constant

```python
def Constant(ls):
  print(ls[0])
Constant([1,23,3])
```

### O(n) linear

```python
def linear(ls):
  for i in ls:
    print(i)
linear([1,23,3])
```

### O(n<sup>2</sup>) Quadratic 

```python
def Quadratic(ls):
  for item1 in ls:
    for item2 in ls:
        print(item1, item2)
Quadratic([1,2,3,4])
```

Many times we are only concerned with the worst possible case of an algorithm but in an interview seeing it is important to keep in mind that worst case and best case scenarios may be completely different big-o times.

```python
def matcher(ls,match):
  for item in ls :
    if item == match:
      return True
  return False
```

```python
matcher([1,2,3,4,5],5) # Worst Case after 5 time index 4 so that Worst case is linear
```

```python
matcher([1,2,3,4,5],1) # Best Case 1 time index 0  so that Best Case is Constant
```

## Dynamic Arrays

Write a function to check whether two given strings are anagram of each other or not. An anagram of a string is another string that contains the same characters, only the order of characters can be different. For example, “abcd” and “dabc” are an anagram of each other.

Note: Ignore space and capital letter

```python
def anagram(s1,s2):
  s1 = s1.replace(' ','').lower()
  s2 = s2.replace(' ','').lower()
  return sorted(s1) == sorted(s2)
```

```python
anagram('dog','God')
```

```python
anagram('mos','God')
```

```python
anagram('clint eastwood','old west action')
```

### Largest Sum⚗

```python
def Largest_SUM(arr):
  if len(arr) == 0:
    return print("Too small")
  max_sum = current_sum = arr[0]
  for num in arr[1:]:
    current_sum = max(current_sum+num,num)
    max_sum = max(current_sum,max_sum)
  return max_sum
```

```python
Largest_SUM([1,7,90,2,60,9,10,30,90,100])
```

### Split a string without using the .split function

```python
def reverse(s):
  length = len(s)
  spaces = [' ']
  words = []
  i = 0
  while i < length:
    if s[i] not in spaces:
      word_start = i 
      while i < length and s[i] not in spaces:
        i+=1
      words.append(s[word_start:i])
    i+=1

  return " ".join(reversed(words))
```

```python
reverse("Mostafa Nabieh")
```

```python
def reverse(s):
  words = []
  current_words=""
  for char in s:
    if char == " ":
      words.append(current_words)
      current_words = ""
    else:
      current_words+=char
  words.append(current_words)

  return " ".join(reversed(words))

```

```python
reverse("Mostafa Nabieh Mourad")
```

## Minesweeper interview question

```python
def mine_weeper(bombs,num_rows,num_cols):
  game = [[0 for i in range(num_cols)]for i in range(num_rows)]
  for bomb_location in bombs:
    (bomb_row,bomb_col) = bomb_location
    game[bomb_row][bomb_col] = -1
    row_range = range(bomb_row-1,bomb_row+2)
    col_range = range(bomb_col-1,bomb_col+2)
    for i in row_range:
      for j in col_range:
        if 0 <= i <num_rows and 0 <= j <num_cols and game[i][j] != -1 :
          game[i][j] += 1 
  return game

mine_weeper([[0,0],[1,2]],3,4)
```

## Frequent Count interview question

```python
def most_frequent(list):
  count = {}
  max_count = 0
  max_item = 0
  for i in list :
    if i not in count:
      count[i] = 1
    else :
      count[i] +=1 
    if count[i] > max_count:
      max_count = count[i]
      max_item = i 
  return max_item
most_frequent([1,2,3,4,10,10,10,6,7,8])
```

## Unique Characters in Strings

```python
def uniq_string(string):
  string = string.replace(' ','')
  return len(set(string)) == len(string)
uniq_string('Ms t f a  n')
```

```python
def uniq_string(string):
  string = string.replace(' ','')
  characters = set()
  for letter in string:
    if letter in characters:
      return False 
    else : 
      characters.add(letter)
  return True
uniq_string('Ms t f a f  n')
```
